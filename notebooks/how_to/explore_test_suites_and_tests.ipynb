# Viewing all available Test Suites and Tests

This notebook aim to demostrate the list of interfaces available to get details of test suites and tests.

## Before you begin

To view Test Suites and Tests in the ValidMind Developer Framework with a Jupyter notebook, you need to install and initialize the client library first.

## Install the client library

```python
%pip install --upgrade -q validmind
```

## Initialize the client library

```python
import validmind as vm
```

## Available test suites

Get the list of test suites available in the ValidMind framework:

```python
vm.test_suites.list_suites()
```

## Test suite details

Get the list of tests available in a given test suite:

```python
vm.test_suites.describe_suite("classifier_full_suite")
```

### Test details

Get the details for a given test:

```python
vm.tests.describe_test('DescriptiveStatistics')
```

### Verbose details view of a test suite and its tests

Get more comprehensive details for test suites and tests:

```python
vm.test_suites.describe_suite("classifier_full_suite", verbose=True)
```

## List all tests

Get the list of tests and their purpose:

```python
vm.tests.list_tests()
```
